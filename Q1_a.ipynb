```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
def Ackley_function_in_2d(x):
    a=20
    b=0.2
    c=2*np.pi
    d=2
    return (-a*(np.exp(-b*(np.sqrt((x[0]*x[0]+x[1]*x[1])/d))))-np.exp((np.cos(c*x[0])+np.cos(c*x[1]))/d)+a+np.exp(1))
```

```python
import numdifftools as nd
```

```python
Ackley_function_in_2d_gradient=nd.Gradient(Ackley_function_in_2d)
```

```python
x =np.arange(-40,40,0.8)
#Plot the function in 3d space by default code
y = np.arange(-40,40,0.8)
x_3d,y_3d =np.meshgrid(x,y)
z_3d = Ackley_function_in_2d([x_3d,y_3d])
 
fig = plt.figure(figsize=(8,8))
 
# syntax for 3-D plotting
ax = plt.axes(projection='3d')
 
# syntax for plotting
ax.plot_surface(x_3d, y_3d, z_3d, cmap='turbo',edgecolor='black')
ax.set_title('Ackley Function')
ax.grid(True)
plt.show()
```

```python

x =np.arange(-1,1,0.005)
#Plot the function in 3d space by default code
y = np.arange(-1,1,0.005)
x_3d,y_3d =np.meshgrid(x,y)
z_3d = Ackley_function_in_2d([x_3d,y_3d])
current_position=(0.61,0.73,Ackley_function_in_2d([0.61,0.73])) 
fig = plt.figure()
learning_rate=.01
# syntax for 3-D plotting

# syntax for plotting
ax = plt.subplot(projection='3d',computed_zorder=False)
X_not=[current_position[0],current_position[1]]
x_path=[]
y_path=[]
z_path=[]
ax.plot_surface(x_3d, y_3d, z_3d, cmap='turbo',edgecolor='none',zorder=0)
for el in range(100):
    Gradi__=Ackley_function_in_2d_gradient(X_not)
    #print(Gradi__)
    X_point,Y_point = current_position[0]-learning_rate*Gradi__[0],current_position[1]-learning_rate * Gradi__[1]
    X_not=[X_point,Y_point]
    current_position =(X_point,Y_point,Ackley_function_in_2d(X_not))
    x_path.append(current_position[0])
    y_path.append(current_position[1])
    z_path.append(current_position[2])
    ax.scatter(X_point,Y_point,Ackley_function_in_2d(X_not),color="black",zorder=1)
    
    # plt.pause(0.001)
    #ax.clear()
ax.set_title('Ackley Function')
plt.show()

```
