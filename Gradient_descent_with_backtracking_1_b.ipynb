```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
def Ackley_function_in_2d(x):
    a=20
    b=0.2
    c=2*np.pi
    d=2
    return (-a*(np.exp(-b*(np.sqrt((x[0]*x[0]+x[1]*x[1])/d))))-np.exp((np.cos(c*x[0])+np.cos(c*x[1]))/d)+a+np.exp(1))
```

```python
import numdifftools as nd
```

```python
Ackley_function_in_2d_gradient=nd.Gradient(Ackley_function_in_2d)
```

```python
x =np.arange(-40,40,0.88)
#Plot the function in 3d space by default code
y = np.arange(-40,40,0.88)
x_3d,y_3d =np.meshgrid(x,y)
z_3d = Ackley_function_in_2d([x_3d,y_3d])
 
fig = plt.figure(figsize=(8,8))
 
# syntax for 3-D plotting
ax = plt.axes(projection='3d')
 
# syntax for plotting
ax.plot_surface(x_3d, y_3d, z_3d, cmap='turbo',edgecolor='black')
ax.set_title('Ackley Function')
ax.grid(True)
plt.show()
```

```python
x_cp=0
y_cp=0
def gradient_descent_with_backtracking(Ackley_function_in_2d,Ackley_function_in_2d_gradient,current_point,alpha=0.1,beta=0.1,rho=0.1,max_iteration=1000,tol=1e-6):
    current_point=[x_cp,y_cp]
    iteration_count=0
    collection=[]
    while iteration_count<max_iteration:
        gradient=Ackley_function_in_2d_gradient(current_point)
        step_size=alpha
        while Ackley_function_in_2d(current_point-step_size*gradient)>Ackley_function_in_2d(current_point)+rho*step_size*np.dot(gradient,gradient):
            step_size*=beta
        current_point= current_point-step_size*gradient
        collection.append(current_point)
        if np.linalg.norm(gradient)<tol:
            break
        iteration_count+=1
    return collection
```

```python
x=np.arange(-1,1,0.005)
y=np.arange(-1,1,0.005)
x_3d,y_3d=np.meshgrid(x,y)
z_3d=Ackley_function_in_2d([x_3d,y_3d])
x_cp=0.61
y_cp=0.73
current_point=[x_cp,y_cp]
point=gradient_descent_with_backtracking(Ackley_function_in_2d,Ackley_function_in_2d_gradient,current_point,0.1,0.1,0.1,1000)
points_3d=[]

#for el in current_point:
 #   a=el
  #  a.append(Ackley_function_in_2d([el[0],el[1]]))
   # points_3d.append([a])
ax=plt.subplot(projection='3d',computed_zorder=False)
ax.plot_surface(x_3d,y_3d,z_3d,cmap='turbo',edgecolor='none',zorder=0)
for el in point:
    ax.scatter(el[0],el[1],Ackley_function_in_2d([el[0],el[1]]),color="black",zorder=1)
    print(el)

plt.show()
```
